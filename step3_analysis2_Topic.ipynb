```python
def hangul_function(text):
    hangul = re.compile('[^ ㄱ-ㅣ 가-힣]')  # 한글 추출 규칙: 띄어 쓰기(1 개)를 포함한 한글
    result = hangul.sub('', text)  # 위에 설정한 "hangul"규칙을 "text"에 적용(.sub)시킴
    return result
```

### 토픽 모델링 완성

```python
temp_list = []


for blog in df['blog_text']:
    # 한글만 남기고, okt를 통해서 명사만 남기는
    LDA_nous = okt.nouns(hangul_function(blog))
    # 2글자 이상의 명사만 남겨서
    LDA_nous = [x for x in LDA_nous if len(x) > 1]  # 한글자 키워드 제거    
    # 불용어 제거
    LDA_nous = [word for word in LDA_nous if not word in stopwords]
    # temp_list에 사용할 명사들만 남긴다.
    temp_list.append(LDA_nous)

## 사용할 명사들을 사전화
dictionary = corpora.Dictionary(temp_list)

# 2개 이상의 문서에 나오고, 50% 이상의 문서에 나오는 경우 제외
dictionary.filter_extremes(no_below=2, no_above=0.5)

# 해당 단어들을 말뭉치로 만들어라
corpus_LDA = [dictionary.doc2bow(text) for text in temp_list]

## 이코드의 의미는 모르겠지만.. ## 빼면 안돌아간다.
temp = dictionary[0]


num_topics = 5 
chunksize = 2000 
passes = 20 
iterations = 400 
eval_every = None # 모델 성능을 평가하는 간격


id2word = dictionary.id2token 

model = LdaModel(
    corpus=corpus_LDA, # 분석에 사용할 말뭉치
    id2word=id2word, # id2word는 단어의 ID와 단어를 매핑하는 딕셔너리
    chunksize=chunksize, # 모델 학습 시 사용되는 문서의 묶음 크기
    alpha='auto',
    eta='auto',
    iterations=iterations,
    num_topics=num_topics, # 토픽 그룹의 갯수를 정해줘
    passes=passes, # 전체 코퍼스를 몇 번 반복하여 학습
    eval_every=eval_every # iteration 횟수
)

lda_visualization = gensimvis.prepare(model, corpus_LDA, dictionary, sort_topics=True)
pyLDAvis.save_html(lda_visualization, './result/topicmodeling_complete.html')
```
