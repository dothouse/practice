```python
# 제주숙박여행 경험이 있는 사람을 기준으로
# 1083 cases

df_jeju = df[df['국내_A_여행횟수_국내숙박_제주'] > 0]
df_jeju
```

```python
# 첫번째 여행 방문지가 제주도 인 경우 1054개
temp1 = df_jeju[(df_jeju['D_TRA1_1_SPOT'] == 39010) | (df_jeju['D_TRA1_1_SPOT'] == 39020)]
temp2 = df_jeju[(df_jeju['D_TRA2_1_SPOT'] == 39010) | (df_jeju['D_TRA2_1_SPOT'] == 39020)]
print('1번째 여행지 제주', len(temp1))
print('2번째 여행지 제주', len(temp2))

# -> 첫번째 여행지가 제주 인사람들로 분석 한정
```

```python
# 제주도 숙박여행 경험이 있는 사람들 중에서
# 첫번째 여행지가 제주도 인 사람들 
df_jeju = df_jeju[(df_jeju['D_TRA1_1_SPOT'] == 39010) | (df_jeju['D_TRA1_1_SPOT'] == 39020)]
df_jeju.head(2)
```
