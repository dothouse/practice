#### Arrival delay의 결측치 처리방법

1. nan을 제거 -> 전체 0.29%, 310건으로 제거하고 간다.
2. 최빈값인 0 
3. 평균값 대체

- 해당 컬럼에서 0과 나머지 숫자는 의미를 갖는다. 0은 지연이 발생하지 않은것, 1이상의 숫자는 지연이 발생된것
-> 2,3으로 적용시에 2번의 경우는 결측치들은 모두 지연이 발생하지 않은것, 3번의 경우는 모두 지연이 발생된것으로 처리된다.

- 숫자의 의미가 상당하고, 결측치의 비율이 적기 때문에 Nan을 분석에서 제외하고, 분석을 진행한다.

```python
## categorical columns 
print(airplane['satisfaction'].unique())
print(airplane['Gender'].unique())
print(airplane['Customer Type'].unique())
print(airplane['Type of Travel'].unique())
```

```python
### label encoder 사용시 앞파벳 순서로 숫자 결정되어서
# apply를 통해 encoding

# 만족도 0,1 -> 만족 1
airplane['satisfaction_bi'] = airplane['satisfaction'].apply(lambda x: 1 if x == 'satisfied' else 0) 

# 'Male - 1' / 'Female -0 '
airplane['Gender_bi'] = airplane['Gender'].apply(lambda x: 1 if x == 'Male' else 0) 

# 'Loyal Customer - 0' / 'disloyal Customer - 0'
airplane['Customer Type_bi'] = airplane['Customer Type'].apply(lambda x: 1 if x == 'Loyal Customer' else 0) 

# Business travel 1 / Personal Travel 0
airplane['Type of Travel_bi'] = airplane['Type of Travel'].apply(lambda x: 1 if x == 'Business travel' else 0) 

airplane.head()
```

```python
airplane.columns
```

```python
# 5점 척도가 아닌 데이터들 정규화

airplane_continous = airplane[['Age', 'Flight Distance', 'Departure Delay in Minutes']]
```

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()
airplane_std = std_scaler.fit_transform(airplane_continous)
airplane_std 
```

```python
airplane_std = pd.DataFrame(airplane_std, columns=['Age_std', 'Flight Distance_std',	'Departure Delay in Minutes_std'])
airplane_std.head()
```

```python
temp = pd.concat([airplane['satisfaction'], airplane.drop(columns='satisfaction'), airplane_std], axis=1)
temp.head()
```

```python
temp.select_dtypes([np.number]).corr()['satisfaction_bi'].sort_values()
```

```python
temp.select_dtypes([np.number]).corr()['satisfaction_bi'].sort_values().plot(kind='bar')
plt.show()
```
