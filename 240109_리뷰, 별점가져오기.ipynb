```python
# 관광지명, 리뷰 가져오기 adsad
result = []
for page in range(1, 3):
    url =f'https://www.visitjeju.net/kr/detail/list?menuId=DOM_000001718000000000&cate1cd=cate0000000002#p{page}&pageSize=12&sortListType=reviewcnt&viewType=map&isShowBtag&tag'
    time.sleep(2)
    driver = webdriver.Chrome()
    driver.get(url)
    time.sleep(5)
    soup = BeautifulSoup(driver.page_source,'html.parser')
    driver.find_element(By.CLASS_NAME,'ico_sort.type_list').click()
    time.sleep(3)

    #  관광지명, 별점 평균 가져오기
    list_a = soup.select('p.s_tit')
    star_mean = soup.select('p.score_count')
    for i in range(len(list_a)):
        text = list_a[i].text
        star =star_mean[0].text.split('\n')[1].lstrip()
        

        # 상세페이지
        driver.find_elements(By.CLASS_NAME,'s_tit')[i].click()
        time.sleep(2)

        #  리뷰 클릭해서 펼치기

        driver.find_element(By.ID,'tab3').click()
        

        # 날짜, 별점 가져오기, (리뷰는 따로 가져오기)
        for page in range(1,11):


            try:
                review_list=driver.find_elements(By.CLASS_NAME,'review_area')
                for re_one in review_list:
                    review = re_one.find_element(By.CLASS_NAME, 'review_txt').text
                    review_date = re_one.find_element(By.CLASS_NAME, 'reg_date').text
                    review_star = re_one.find_element(By.CLASS_NAME, 'score_area_p').text
                    

            except:
                review = 'error'
                review_date = 'error'
                review_star = 'error'
                    
                result.append([text,star,review,review_date, review_star])
            

            time.sleep(2)

            try:
                driver.find_element(By.CLASS_NAME,'spr_com.page-next').click()
                time.sleep(2)
            except:
                pass

            
            time.sleep(2)
        
        
            
        driver.back()
        time.sleep(0.3) 
        driver.back() 
    
```

```python
# 관광지명, 리뷰 가져오기 adsad
result = []
for page in range(120, 125):
    url =f'https://www.visitjeju.net/kr/detail/list?menuId=DOM_000001718000000000&cate1cd=cate0000000002#p{page}&pageSize=12&sortListType=reviewcnt&viewType=map&isShowBtag&tag'
    driver = webdriver.Chrome()
    driver.get(url)
    driver.implicitly_wait(3)
    
    driver.find_element(By.CLASS_NAME,'ico_sort.type_list').click()
    driver.implicitly_wait(3)
    time.sleep(1)

    soup = BeautifulSoup(driver.page_source, 'html.parser')
    

    #  관광지명, 별점 평균 가져오기
    lis = soup.select('.item_section')

    if len(lis) != 12:
        break
    
    cnt = 0
    for li in lis[:12]:
        title = li.select_one('.s_tit').text.strip()
        tag = li.select_one('.item_tag').text.strip()
        
        try :
            star = int(re.sub('[^0-9]','',li.select_one('p.score_count').text)[-1])
        except :
            star = 0
        
        
        # 상세페이지        
        driver.find_elements(By.CLASS_NAME, 'item_top')[cnt].click()
        cnt += 1
        time.sleep(round(random.uniform(0.8, 1.5), 3))

        link = driver.current_url

        # 리뷰 클릭해서 펼치기
        driver.find_element(By.ID,'tab3').click()

        # 리뷰 갯수로 리뷰 페이지 갯수
        review_num = driver.find_element(By.ID, 'tab3').find_element(By.CLASS_NAME, 'cont_tit').text
        review_num = int(re.sub('[^0-9]','',review_num))
        
        if review_num > 100:
            review_page = 19
        elif review_num == 0:
            review_page = 0
        else :
            review_page = int((review_num - 1)// 5)
    

        if review_num == 0:    
            review_text = 'none'
            review_date = 'none'
            review_star = 'none'
            review_tag = 'none'                
            result.append([title, star, tag, link, review_text, review_date, review_star, review_num, review_tag])  

        else :
            for i in range(review_page + 1):
                review_list = driver.find_elements(By.CLASS_NAME, 'review_area')
                for review in review_list:
                    review_text = review.find_element(By.CLASS_NAME, 'review > p' ).text.strip()
                    review_date = review.find_element(By.CLASS_NAME, 'reg_date').text.strip()
                    try :
                        review_star = review.find_element(By.CLASS_NAME, 'score_count_p').text.strip()
                        review_star = int(re.sub('[^0-9]','',review_star)[-1])
                    except :
                        review_star = 0
                        
                    review_tag = []
                    review_tag_list = review.find_element(By.CLASS_NAME, 'check_list').find_elements(By.CLASS_NAME, 'top')
                    for review_tag_li in review_tag_list:
                        tag_temp = review_tag_li.text.strip()             
                        review_tag.append(tag_temp)

                    result.append([title, star, tag, link, review_num, review_text, review_date, review_star, review_tag])

                if review_num > 5:
                    driver.find_element(By.CLASS_NAME,'spr_com.page-next').click()
                    time.sleep(round(random.uniform(0.9, 1.5), 3))                    
                

        
        time.sleep(round(random.uniform(0.9, 1.5), 3))   
            
        driver.back()
        time.sleep(0.3) 
        driver.back()

```
