```python
import os
import json
import pandas as pd

# Define the folder path
base_path = 'D:/song/dataset/요약문 및 레포트 생성 데이터/01.데이터/1.Training/라벨링데이터/'

# Initialize an empty list to store JSON data from all files
subfolder1 = [f.path + '/20per' for f in os.scandir(base_path) if f.is_dir()]
subfolder2 = [f.path + '/2~3sent' for f in os.scandir(base_path) if f.is_dir()]

subfolders = subfolder1 + subfolder2


# Initialize an empty list to store JSON data from all files
all_json_data = []

# Iterate through all files in the folder
for path in subfolders:
    for file_name in os.listdir(path ):
        # Check if the file is a JSON file
        if file_name.endswith('.json'):
            # Construct the full file path
            file_path = os.path.join(path, file_name)
            # Open the file and load JSON data
            with open(file_path, 'r', encoding='utf-8') as f:
                json_data = json.load(f)
                all_json_data.append(json_data)
```

```python
raw = pd.DataFrame(all_json_data)
raw
```

```python
cols1 = raw['Meta(Acqusition)'][0].keys()

result1 = []

for i in range(len(raw)):
    temp = []
    for col in cols1:         
        temp.append(raw['Meta(Acqusition)'][i][col])

    result1.append(temp)
```

```python
df1 = pd.DataFrame(result1, columns=cols1)
print(len(df1))
df1.head()
```

```python
cols2 = raw['Meta(Refine)'][0].keys()

result2 = []

for i in range(len(raw)):
    temp = []
    for col in cols2:         
        temp.append(raw['Meta(Refine)'][i][col])

    result2.append(temp)
```

```python
df2 = pd.DataFrame(result2, columns=cols2)
print(len(df2))
df2.head()
```

```python
cols3 = raw['Annotation'][0].keys()

result3 = []

for i in range(len(raw)):
    temp = []
    for col in cols3:         
        temp.append(raw['Annotation'][i][col])

    result3.append(temp)
```

```python
df3 = pd.DataFrame(result3, columns=cols3)
print(len(df3))
df3.head()
```

```python
df = pd.concat([df1, df2, df3], axis=1)
df.head()
```

```python
for i in range(3):
    print(df['passage'][i])
    print('----------------')
    print(df['summary1'][i])
    print('----------------')
    print(df['summary2'][i])
    print('----------------')
    print(df['summary3'][i])
    print('----------------')
```

```python
df.to_csv('D:/song/dataset/요약문 및 레포트 생성 데이터/summary.csv', index = False)
```
