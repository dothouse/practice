```python
import os
import json
import pandas as pd
```

```python

# Define the folder path
base_path = 'D:/song/dataset/book/Training/'

# List all directories (folders) within the specified folder
subfolders = [f.path for f in os.scandir(base_path) if f.is_dir()]

# Initialize an empty list to store JSON data from all files
all_json_data = []

# Iterate through all files in the folder
for path in subfolders:
    for file_name in os.listdir(path):
        # Check if the file is a JSON file
        if file_name.endswith('.json'):
            # Construct the full file path
            file_path = os.path.join(path, file_name)
            # Open the file and load JSON data
            with open(file_path, 'r', encoding='utf-8') as f:
                json_data = json.load(f)
                all_json_data.append(json_data)
```

```python
df_base = pd.DataFrame(all_json_data)
df_base
```

```python
cols = df_base['info'][0].keys()

result = []

for i in range(len(all_json_data)):
    temp = []
    for col in cols:         
        temp.append(pd.DataFrame(all_json_data)['info'][i][col])
    result.append(temp)
df_meta = pd.DataFrame(result, columns=cols)
```
