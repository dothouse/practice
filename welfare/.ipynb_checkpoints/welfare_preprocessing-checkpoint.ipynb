### marriage 변수
- 0 - 비해당 -> NA로
- 1 - 유배우 -> 혼인상태(1)
- 2 - 사별, 3 - 이혼, 4 - 별거 -> 기타(2)
- 5 - 미혼 (0)
- 6 - 기타 -> 기타(2)


````python
df.marriage.unique()
````

````python
def marriage_recode(x):
    if x == 0:
        temp = np.nan
    elif x == 1:
        temp = 1
    elif (x == 2) | (x == 3) | (x == 4) | (x == 6):
        temp = 2
    elif (x == 5):
        temp = 0
    
    return temp

df['marriage2'] = df['marriage'].apply(marriage_recode)
df.head(5)
````

### 소득
- 분석이 용이성을 위해 3개의 소득만 가져옴
- 상용(regular), 임시직(temp), 자영업(owner)
- 3개 변수를 결합해서 사용
- 8258개의 NaN값
- 음수값 - nan으로
- 이상치제거

````python
df['inc'] = df.apply(lambda x: np.nan if (pd.isna(x.inc_regular)) & (pd.isna(x.inc_temp))
                      & (pd.isna(x.inc_owner))
                      else x[['inc_regular', 'inc_temp', 'inc_owner']].sum(),  axis = 1)
````

````python
df['inc'].isna().sum()
````

````python
# inc <= 0 인 경우 nan 값으로 처리
df['inc'] = df['inc'].apply(lambda x: np.nan if x <=0 else x)
df['inc']
````

````python
# 8258 -> 8307로 증가
df['inc'].isna().sum()
````

````python
# inc 이상치 확인 및 제거
# Q3 + 1.5 * IQR > inc
# Q1 - 1.5 * IQR < inc
# 위 구간을 벗어날시에 이상치로 판단하고 제거
# 293개가 이상치로 판단됨
inc_IQR = df.inc.quantile(0.75) - df.inc.quantile(0.25)
inc_Q3 = df.inc.quantile(0.75)
inc_Q1 = df.inc.quantile(0.25)

df_inc_outlier = df.inc[(df.inc > (inc_Q3 + (1.5 * inc_IQR))) | (df.inc < (inc_Q1 - (1.5 * inc_IQR)))]
len(df_inc_outlier)
````

````python
df['inc'] = df['inc'].apply(lambda x: np.nan if (x > (inc_Q3 + 1.5 * inc_IQR)) or (x < inc_Q1 - (1.5 * inc_IQR)) else x)
df['inc'].isna().sum()
# nan 8307 -> 8597 으로 변화
````

````python
df['inc_log'] = df['inc'].apply(lambda x: np.log(x) if x >= 1 else x)
df['inc_log']
````
