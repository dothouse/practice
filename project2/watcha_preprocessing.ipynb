```python
import pandas as pd
import re
```

## 년도별 파일 결합

```python
info1 = pd.read_csv('data/watcha/movie_info_2023_1.csv')
info2 = pd.read_csv('data/watcha/movie_info_2023_2.csv')
```

```python
df_info = pd.concat([info1, info2])
df_info.reset_index(drop=True, inplace=True)
df_info
```

```python
review1 = pd.read_csv('data/watcha/movie_review_2023_1.csv')
review2 = pd.read_csv('data/watcha/movie_review_2023_2.csv')
```

```python
df_review = pd.concat([review1, review2])
df_review.reset_index(drop=True, inplace=True)
df_review
```

## df_info 전처리

```python
df_info.head()
```

```python
# info1 정보 (제작년도, 장르, 국적)
df_info1 = df_info['info1'].str.split('·', expand=True)
df_info1.columns = ['making_year', 'genre', 'nation']
df_info1.head()
```

```python
# 장르 정보 변환
df_genre = df_info1['genre'].str.split('/', expand=True)
df_genre.columns = ['genre' + str(x) for x in range(1, len(df_genre.columns) + 1)]
df_genre
```

```python
df_info2 = df_info['info2'].str.split('·', expand=True)
df_info2.columns = ['running_time', 'film_ratings']
df_info2.head()
```

```python
df_info = pd.concat([df_info, df_info1, df_genre, df_info2], axis=1).drop(columns=['info1', 'info2'])
df_info.head()
```

```python
df_info['rating_cnt']
```

```python
def clean_rating_cnt(value):
    cleaned_value = re.sub('[()명,.]', '', value)
    if '만' in cleaned_value:
        return int(cleaned_value.replace('만', '')) * 1000
    return int(cleaned_value)
```

```python
df_info['rating_cnt'] = df_info['rating_cnt'].apply(clean_rating_cnt)
df_info['rating_cnt'].head(10)
```

```python
df_info.head()
```

```python
df_info.to_csv('data/watcha/movie_info.csv', index=False)
```

## 출연배우 정보 정리 
- 새로운 df으로 분리

```python
df_info['actor_list'] = df_info['actor_list'].apply(lambda x: eval(x))
df_info['actor_list']
```

```python
for i in range(len(df_info['actor_list'][0])):
    print(df_info['actor_list'][0][i])
```

```python
# 역할에 따라 컬럼을 나누기
df_info['pd'] = df_info['actor_list'].apply(lambda x: [actor for actor in x if '감독' in actor])
df_info['main_actors'] = df_info['actor_list'].apply(lambda x: [actor for actor in x if '주연' in actor])
df_info['supporting_actors'] = df_info['actor_list'].apply(lambda x: [actor for actor in x if '조연' in actor])
df_info['actors'] = df_info['actor_list'].apply(lambda x: [actor for actor in x if ('출연' in actor) & ('특별출연' not in actor)])
```

```python
df_info.head()
```

```python
# 주연정보
df_main_actor = df_info['main_actors'].apply(lambda x: pd.Series([actor.split('(주연')[0] for actor in x]))
df_main_actor.columns = ['main_actor' + str(x) for x in range(1, len(df_main_actor.columns) + 1)]
df_main_actor.head()
```

```python
# 조연정보
df_supporting_actors = df_info['supporting_actors'].apply(lambda x: pd.Series([actor.split('(조연')[0] for actor in x]))
df_supporting_actors.columns = ['supporting_actors' + str(x) for x in range(1, len(df_supporting_actors.columns) + 1)]
df_supporting_actors.head()
```

```python
# 감독정보
df_pd = df_info['pd'].apply(lambda x: pd.Series([actor.split('(감독')[0] for actor in x]))
df_pd.columns = ['pd' + str(x) for x in range(1, len(df_pd.columns) + 1)]
df_pd.head()
```

```python
#  주연/조연인 아닌 출연이라고 적혀있는 경우
df_actors = df_info['actors'].apply(lambda x: pd.Series([actor.split('(출연')[0] for actor in x]))
df_actors.columns = ['actors' + str(x) for x in range(1, len(df_actors.columns) + 1)]
df_actors.head()
```

```python
df_actor = pd.concat([df_info['title'], df_pd, df_main_actor, df_supporting_actors, df_actors], axis=1)
df_actor.head()
```

```python
# df_info = df_info.drop(columns=['pd', 'main_actors', 'supporting_actors', 'actors'])
# df_info.head()
```

```python
df_actor.to_csv('data/watcha/movie_actor.csv', index=False)
```
