### 제주도 맛집 목록
- 출처 : kakao map(크롤링)
- 데이터 수집 : 김제이
- 식당 매출 데이터(내지인 / 외지인 비율)을 통해 2000건 샘플링하고, 해당 식당 정보 수집

```python
df_food = pd.read_excel('d://song/project1_practice/data/final/제주도 식당 목록.xlsx')      
df_food.head()
```

```python
df_food =df_food[['이름','주소','주소2','위도','경도','구분','구분2']]
df_food.drop_duplicates(subset='이름', inplace=True)
df_food.head(2)
```

```python
df_food.reset_index(drop = True, inplace=True)
df_food.reset_index(inplace=True)
df_food.columns = ['id', 'name',  'addr' ,  'dong' ,'lat', 'lng' , 'detailtype_str', 'detailtype2_str']
df_food.head(2)
```

```python
df_food['detailtype_str'].unique()
```

```python
# 아시아 음식 - 일식, 중식 ->1
# 양식 - 기타
# 주점 및 주류 판매
# 카페 - 음료 간식,베이커리, 디저트
# 패스트푸드 - 피자 치킨 햄버거
# 한식 - 분식, 고기
```

```python
df_food['detailtype'] = df_food['detailtype_str'].apply(lambda x: 1 if x in ['일식.회','일식','회', '중식', '아시아음식'] else 
                                                    (2 if x in ['기타', '양식'] else 
                                                     (3 if x in ['주점및주류판매', '주점 및 주류판매', '주점'] else 
                                                      (4 if x in ['음료', '간식', '베이커리.디저트', '카페'] else
                                                        (5 if x in ['피자.치킨.햄버거', '패스트푸드'] else
                                                          (6 if x in ['한식', '분식', '고기'] else 0))))))

```

```python
df_food.groupby('detailtype').size().reset_index(name='count')
```

```python
df_food
```

```python
df_food = df_food[['id', 'name',  'addr','lat', 'lng',  'dong','detailtype',  'detailtype_str', 'detailtype2_str']]
df_food.head(2)
```

```python
df_food.to_csv('d://song/project1_practice/data/final/food.csv', index = False)  
```
