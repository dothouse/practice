```python
import pandas as pd
import numpy as np
```

### step1 - 기초

```python
population = pd.read_csv('d:/song/python_practice/train/pop.csv')
df_pop = population.copy()
df_pop.head(10)
```

```python
df_pop.drop(index = [0,1], inplace = True)
df_pop.head(10)
```

```python
df_pop.columns = ['name', 'pop_2016', 'pop_2017', 'pop_2018', 'pop_2019',
                         'pop_2020', 'pop_2021', 'pop_2022']
df_pop
```

### step2 - 시군구 지역 명칭 정리
- 중구, 남구 등 중복된 명칭이 존재, 이를 해결하기 위해서 
- 시도 + 시군구를 통해서 city2 변수 생성

```python
test = df_pop.apply(lambda x: x['name'] if '특별' in x['name'] 
                    else x['name']  if '북도' in x['name'] 
                    else x['name']  if '남도' in x['name'] 
                    else x['name']  if '광역' in x['name']    
                    else x['name']  if '경기도' in x['name'] 
                    else np.nan, axis = 1)
test.unique()
```

```python
df_pop['sido'] = df_pop.apply(lambda x: x['name'] if '특별' in x['name'] 
                              else x['name']  if '북도' in x['name'] 
                              else x['name']  if '남도' in x['name'] 
                              else x['name']  if '광역' in x['name']    
                              else x['name']  if '경기도' in x['name'] 
                              else np.nan, axis = 1)
df_pop['sido'].ffill(inplace = True)
df_pop
```

```python
# sido 와 name이 같은 경우 시도 전체의 데이터
df_pop['city2'] = df_pop.apply(lambda x: x['sido'] +' '+  x['name'] if x['sido'] != x['name'] else np.nan, axis = 1)
df_pop = df_pop[df_pop['city2'].isna() == False]
df_pop
```

### step3 - 데이터 정리
- 값이 존재하지 않는 row ( '-' ) 제거 후 전부 숫자행으로 변경
- 인천광역시 남구 -> 인천광역시 미추홀구


```python
df_pop = df_pop.map(lambda x: np.nan if x == '-' else x) #df의 전체를 바꿀땐 map함수로
df_pop = df_pop.apply(lambda x: pd.to_numeric(x, errors='coerce').astype('float'))
df_pop.info()
```

```python
df_pop[(df_pop['city2'] == '인천광역시 남구')|(df_pop['city2'] == '인천광역시 미추홀구')].sum(axis = 0, numeric_only=True)
```
