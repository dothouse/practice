```python
import pandas as pd
import numpy as np
```

### step1 - 기초

```python
migration = pd.read_csv('d:/song/python_practice/train/migration.csv')
df_mig = migration.copy()
df_mig.head(10)
```

```python
df_mig.columns = ['name', 'etc', 
                  'in_2016', 'out_2016', 'migration_2016',
                  'in_2017', 'out_2017', 'migration_2017',
                  'in_2018', 'out_2018', 'migration_2018',
                  'in_2019', 'out_2019', 'migration_2019',
                  'in_2020', 'out_2020', 'migration_2020',
                  'in_2021', 'out_2021', 'migration_2021',
                  'in_2022', 'out_2022', 'migration_2022']
                  
df_mig
```

```python
df_mig.drop(index = [0,1], inplace = True)
df_mig.drop(columns = 'etc', inplace = True)
df_mig.head(10)
```

### step2 - 시군구 지역 명칭 정리
- 중구, 남구 등 중복된 명칭이 존재, 이를 해결하기 위해서 
- 시도 + 시군구를 통해서 city2 변수 생성

```python
test = df_mig.apply(lambda x: x['name'] if '특별' in x['name'] 
                    else x['name']  if '북도' in x['name'] 
                    else x['name']  if '남도' in x['name'] 
                    else x['name']  if '광역' in x['name']    
                    else x['name']  if '경기도' in x['name'] 
                    else np.nan, axis = 1)
test.unique()
```

```python
df_mig['sido'] = df_mig.apply(lambda x: x['name'] if '특별' in x['name'] 
                              else x['name']  if '북도' in x['name'] 
                              else x['name']  if '남도' in x['name'] 
                              else x['name']  if '광역' in x['name']    
                              else x['name']  if '경기도' in x['name'] 
                              else np.nan, axis = 1)
df_mig['sido'].ffill(inplace = True)
df_mig
```

```python
# sido 와 name이 같은 경우 시도 전체의 데이터
df_mig['city2'] = df_mig.apply(lambda x: x['sido'] +' '+  x['name'] if x['sido'] != x['name'] else np.nan, axis = 1)
df_mig = df_mig[df_mig['city2'].isna() == False]
df_mig
```
