```python
temp = pd.DataFrame()
station = []
road_addr = []
for i in range(1, 23):
    if i < 22 :
        for j in range(2, 12):
            url = f'https://www.kric.go.kr/jsp/board/portal/sub05/est/estationList.jsp?pageNo={i}'
            resp = requests.get(url)
            html = resp.text
            soup = BeautifulSoup(html, 'html.parser')
            station_name = soup.select('.bd_railway_station')[j].text
            station_roadaddr = soup.select('.bd_railway_add')[j].text
            station.append(station_name)
            road_addr.append(station_roadaddr)
    # 22page는 5개역만 존재
    else :
        for j in range(2, 7):
            url = f'https://www.kric.go.kr/jsp/board/portal/sub05/est/estationList.jsp?pageNo={i}'
            resp = requests.get(url)
            html = resp.text
            soup = BeautifulSoup(html, 'html.parser')
            station_name = soup.select('.bd_railway_station')[j].text
            station_roadaddr = soup.select('.bd_railway_add')[j].text
            station.append(station_name)
            road_addr.append(station_roadaddr)
        

temp['station2'] = station
temp['road_addr'] = road_addr
temp
```

```python
temp['station2'].unique()
```

```python
# 이름에 예외가 있는 역들 수정
temp['station2'] = temp['station2'].str.replace('김천(구미)','김천구미역')
temp['station2'] = temp['station2'].str.replace('진부(오대산)역','진부역')
temp['station2'] = temp['station2'].str.replace('양원(영동선)역','양원역')
temp['station2']
```

```python
temp['station2'].unique()
```

```python
addr = pd.merge(addr, temp,
         how = 'left',
         on = 'station2')
addr
```

```python
import time

def find_addr(station_name):
    base_url = 'https://www.juso.go.kr/addrlink/addrLinkApi.do'
    params1 = f'confmKey={addr_key}&currentPage=1&countPerPage=10'
    params2 = f'keyword={quote(station_name)}&resultType=json'
    url = f'{base_url}?{params1}&{params2}'
    result = requests.get(url)
    try:
        if result.status_code == 200:
            res = json.loads(result.text)
            return result.json()['results']['juso'][0]['roadAddr']
        else :
            return 'none'
    except:
        return 'what'
    time.sleep(0.3)
```

```python
addr['road_addr2'] = addr['station2'].apply(lambda x: find_addr(x))
```

```python
addr['road_addr'] = addr.apply(lambda x: x['road_addr1'] if not pd.isna(x['road_addr1']) else x['road_addr2'], axis=1)
addr
```

### step3 - KAKAO api로 좌표 획득하기

```python
# api 가져오기
with open('D:/song/class_appendix/key/kakao_map_key.txt') as file:
    kakao_key = file.read()
```

```python
# https://developers.kakao.com/docs/latest/ko/local/dev-guide#address-coord
base_url = 'https://dapi.kakao.com/v2/local/search/address.json'
addr_test = '서울특별시 서대문구 신촌역로 4 (대현동)'
url = f'{base_url}?query={quote(addr_test)}'
# header : Authorization: KakaoAK ${REST_API_KEY}
header = {'Authorization': f'KakaoAK {kakao_key}'}
```

```python
result = requests.get(url, headers = header).json()
result
```

```python
result['documents'][0]['address']['x'] #lon
```

```python
result['documents'][0]['address']['y'] #lat
```

```python
def kakao_api_lat(road_addr):
    base_url = 'https://dapi.kakao.com/v2/local/search/address.json'
    addr_input = road_addr
    url = f'{base_url}?query={quote(addr_input)}'
    header = {'Authorization': f'KakaoAK {kakao_key}'}
    result = requests.get(url, headers = header).json()
    try:
        return result['documents'][0]['address']['y']
        
    except:
        return 'what'
    time.sleep(0.3)
   
```

```python
def kakao_api_lon(road_addr):
    base_url = 'https://dapi.kakao.com/v2/local/search/address.json'
    addr_input = road_addr
    url = f'{base_url}?query={quote(addr_input)}'
    header = {'Authorization': f'KakaoAK {kakao_key}'}
    result = requests.get(url, headers = header).json()
    try:
        return result['documents'][0]['address']['x']
        
    except:
        return 'what'
    time.sleep(0.3)
```

```python
addr['lat'] = addr['road_addr'].apply(lambda x: kakao_api_lat(x))
addr['lon'] = addr['road_addr'].apply(lambda x: kakao_api_lon(x))
addr
```
