### 결합데이터 정리

```python
food = pd.concat([food1, food2], axis = 0)
food.reset_index(drop = True, inplace = True)
food
```

```python
food.columns
```

```python
# subset some columns
food = food[['번호','인허가일자', '인허가취소일자', '영업상태구분코드', '영업상태명', '상세영업상태코드', '폐업일자',
             '소재지전체주소', '도로명전체주소', '사업장명', '업태구분명', '좌표정보(x)', '좌표정보(y)', '시설총규모']]
food.head(2)
```

```python
# address 정리
food.loc[:, 'sido'] = food['소재지전체주소'].apply(lambda x: x.split()[0]  if isinstance(x, str) else None)
food.loc[:, 'city'] = food['소재지전체주소'].apply(lambda x: x.split()[1]  if isinstance(x, str) and len(x.split()) > 1 else None)
food.loc[:, 'dong']  = food['소재지전체주소'].apply(lambda x: x.split()[2]  if isinstance(x, str) and len(x.split()) > 2 else None)
food.loc[:, 'city2'] = food['소재지전체주소'].apply(lambda x: x.split()[0] + " " +x.split()[1] 
                                             if isinstance(x, str) and len(x.split()) > 2 else None)
food[['소재지전체주소', 'sido', 'city', 'dong', 'city2']]
```

```python
def coord_trans(coord, coord_orgin, coord_trans):
    y_, x_ = Transformer.from_crs(coord_origin_type, coord_trans_type).transform(
            coord[:,0], coord[:, 1])
    temp = np.dstack([y_, x_])[0]
    return pd.DataFrame(temp, columns=['lat', 'lon'])
    
```
