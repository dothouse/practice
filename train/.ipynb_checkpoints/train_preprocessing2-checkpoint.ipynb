```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('D:/song/python_practice/0data/train/train.csv')
df = train.copy()
df.head(10)
```

```python
df = df[df['station'] != '합계']
df['station'] = df['station'].replace("서대구('22.3.31~)", '서대구')
df

for i in range(2016, 2023):
    df = df[df[f'on_total_{i}'].isna() == False]
```

```python
# value가 '-'인경우 존재
df.iloc[:,21]
```

```python
# 열이 object인 경우 
# 빈칸 제거, - -> 0으로
# company, station은 error 발생해서 ...
for i in df.columns[2:]:
    if df[i].dtype == 'O':
        df[i] = df[i].str.strip().replace('-', 0)
    df[i] = pd.to_numeric(df[i], errors='coerce')
```

```python
df.iloc[:,21]
```

```python
#  on_total이 1개라도 없는 기차역는 제외
for i in range(2016, 2023):
    df = df[df[f'on_total_{i}'].isna() == False]

df.reset_index(drop = True, inplace = True) # index 다시 붙이기
df
```

```python
# 총 196개의 기차역이 선정됨
df.station.count(), df.station.nunique() 
```

### korail과 sr 동시 운영기차역 처리방법
- 총 기차역 210 / unique 196
- 총 14개의 기차역이 동시운영됨<br>
->이 부분을 어떻게 처리할까?

```python
dup_station = df['station'].value_counts().to_frame()
dup_station_list = list(dup_station[dup_station['count'] > 1].index)
dup_station_list
```

```python
#컬럼만 있는 데이터 프레임 만들기 
temp = df.iloc[0,].to_frame().T.drop(index = 0)
```

```python
for i in dup_station_list:
    temp = pd.concat([temp, df[df.station == i].sum(axis = 0).to_frame().T])
```

```python
temp['company'] = 'total'
temp['station'] = dup_station_list
```

```python
temp
```

```python
# 중복역 리스트에 있는 역 제거 
for i in dup_station_list:
    df = df[df['station'] != i]
```

```python
# 중복역 계산 데이터 프레임 temp와 중복역이 제거된 df 결합
df = pd.concat([df, temp])
df.drop(columns = 'company', inplace = True)
df
```

```python
df.to_csv('D:/song/python_practice/0data/train/train.csv')
```
