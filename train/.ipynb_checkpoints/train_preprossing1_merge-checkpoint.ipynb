```python
# 컬럼명 변경
for i in range(2016, 2023):
    df_name = globals()['train' + '_' + str(i) + '_7']
    
    if i < 2018:
        df_name.drop(columns = ['Unnamed: 2', 'Unnamed: 3'], inplace = True)
        
    df_name.columns =['company', 'station', f'total_{i}',
                      f'on_ktx_{i}', f'on_saemaul_{i}', f'on_mugungwha_{i}',
                      f'on_commute_{i}', f'on_total_{i}',
                      f'off_ktx_{i}', f'off_saemaul_{i}', f'off_mugungwha_{i}',
                      f'off_commute_{i}', f'off_total_{i}']
    df_name.drop(df_name.index[0:5], inplace = True)
    df_name['company'].fillna(method = 'ffill', inplace = True)
    df_name['company'] = df_name['company'].apply(lambda x : 'korail' if (x == '한국철도공사\nKorail') |  (x == '한국철도공사')
                                                  else 'sr'
                                                  if x == '㈜SR' else x )
    # 역명 양쪽에에 공백 존재하는 sheet 존재 
    df_name['station'] = df_name['station'].str.strip()
    
    df_name.dropna(subset = ['station'], inplace = True)
```
