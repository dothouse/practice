```python
import pandas as pd
```

# step1 - sheet 7 - 역별 여객 수송실적 정리

```python
# 철도통계연보 - 지역간 철도 2022
# sheet 7 - 역별 여객 수송실적
train_2022_s6 = pd.read_excel('D:/song/python_practice/0data/train/2022/4. 수송(여객)_완.xlsx',
                              sheet_name = 6)

train_2022_s6.head()
```

### 컬럼명 변경

```python
train_2022_s6.columns = ['company', 'station', 'total_2022',
                         'on_ktx_2022', 'on_saemaul_2022', 'on_mugungwha_2022', 'on_commute_2022', 'on_total_2022',
                         'off_ktx_2022', 'off_saemaul_2022', 'off_mugungwha_2022', 'off_commute_2022', 'off_total_2022']
train_2022_s6.head(10) 
```

```python
# 필요없는 행/열 제거
train_2022_s6.drop(train_2022_s6.index[0:6], inplace = True)
train_2022_s6.drop(columns = 'company', inplace = True)
train_2022_s6 = train_2022_s6[train_2022_s6['station'].isna() == False] # 역명 NaN 제거
train_2022_s6.head(5)
```

### step 2 - sheet 8. 역별여객승하차실적 (총계) 정리

```python
# 철도통계연보 - 지역간 철도 2022
# sheet 8 - 8. 역별여객승하차실적 (총계)

train_2022_s7 = pd.read_excel('D:/song/python_practice/0data/train/2022/4. 수송(여객)_완.xlsx',
                    sheet_name = 7)

train_2022_s7.head()
```

```python
train_2022_s7.tail(20)
```

### 컬럼명 변경
- 0 운영사 1 역명 2-5 하행관련, 6-9 상행관련

```python
train_2022_s7.columns = ['company', 'station',
                         'down_on_2022', 'down_off_2022', 'down_total_2022','down_kilo_2022',
                         'up_on_2022', 'up_off_2022', 'up_total_2022', 'up_kilo_2022']

train_2022_s7.head()
```

### 필요없는 열과 행 제거


```python
train_2022_s7 = train_2022_s7[train_2022_s7['station'].isna() == False] # 역명 NaN 제거
train_2022_s7.drop(columns = 'company', inplace = True)
train_2022_s7.head()
```

### step3 - s7에 s6을 merge
- key = station

```python
train_2022 = train_2022_s7.merge(train_2022_s6, 
                                 on = 'station',
                                 how = 'left')

train_2022.station.unique(), train_2022.station.nunique()
```
