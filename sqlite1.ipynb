#### SQLite


```python
from sqlalchemy import create_engine, MetaData, Table, Column, Integer, String

import pandas as pd
import numpy as np 
```

```python
engine = create_engine('sqlite:///test.db', echo = True)
meta = MetaData()

company = Table(
   'test', meta, 
   Column('id', Integer, primary_key = True), 
   Column('reviewNum', Integer), 
   Column('addr', String),
   Column('title', String)
)

meta.create_all(engine)
```

```python
df = pd.read_excel('./data/숙소주소.xlsx')
df.head(3)
```

```python
df = df[df['주소1'] != 'none']
df.head(3)
```

```python
df.rename(columns={'Unnamed: 0' : 'id'}, inplace=True)
df.reset_index(drop= True, inplace=True)
df.head(3)
```

```python
data = []
for i in range(len(df)):
    reviewNum = df['reviewNum'][i]    
    addr = df['주소1'][i]
    title = df['title'][i]
    data.append({'reviewNum' : reviewNum, 'addr' : addr, 'title' : title})
```

```python
data
```

```python
conn = engine.connect()
result = conn.execute(company.insert(), data)
```
