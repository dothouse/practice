```python
import pandas as pd
```

```python
df = pd.read_excel('./data/place.xlsx')
df
```

```python
df.drop(columns=df.columns[2:14], inplace= True)
df.drop(columns=df.columns[0], inplace= True)
df
```

```python
col_names = ['ID']
for i in range(len(df.columns) - 1):
    col_names.append(f'pension{i}')

col_names   
```

```python
df.columns = col_names
```

```python
df['type'] = df['ID'].apply(lambda x: x // 10000)
df.head(5)
```

```python
result = []
for i in df.columns[1:-2]:
    for spot_type in range(1, 7):
        temp = df[df['type'] == spot_type]
        cnt1 = ((temp[f'{i}'] < 1).sum())
        cnt2 = ((temp[f'{i}'] < 2).sum())
        cnt3 = ((temp[f'{i}'] < 3).sum())
        cnt5 = ((temp[f'{i}'] < 5).sum())
        cnt10 = ((temp[f'{i}'] < 10).sum())
        cnt15 = ((temp[f'{i}'] < 15).sum())
        cnt20 = ((temp[f'{i}'] < 20).sum())
        result.append([i, spot_type, cnt1, cnt2, cnt3, cnt5, cnt10, cnt15, cnt20])
```

```python
df_result = pd.DataFrame(result)
df_result.columns = ['name', 'type', 'cnt_1km','cnt_2km', 'cnt_3km','cnt_5km', 'cnt_10km', 'cnt_15km', 'cnt_20km']
```
